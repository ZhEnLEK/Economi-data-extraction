```python
import sqlite3
from sqlite3 import Error

#Create table

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn


def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)


def main():
    database = r"C:\Users\zhenl\Desktop\New Shiny App\new\economic.db"

    sql_create_xt_table = """ CREATE TABLE IF NOT EXISTS ExternalTrade (
                                        ID INTEGER PRIMARY KEY,
                                        Indicator_ID smallint NOT NULL,
                                        Trade_ID smallint NOT NULL,
                                        Country_ID smallint NOT NULL,
                                        Year smallint NOT NULL,
                                        Month smallint NOT NULL,
                                        Value_RM_millions float NOT NULL,
                                        Is_percentage bit NOT NULL,
                                        Is_M2M bit NOT NULL,
                                        Date_Created datetime,
                                        Created_By smallint NOT NULL,
                                        Date_Modified datetime,
                                        Modified_By,
                                        Is_Active bit NOT NULL,
                                        Is_Delete bit NOT NULL
                                        
                                    ); """

   
    # create a database connection
    conn = create_connection(database)

    # create tables
    if conn is not None:
        # create projects table
        create_table(conn, sql_create_xt_table)

        # create tasks table
       # create_table(conn, sql_create_tasks_table)
    else:
        print("Error! cannot create the database connection.")


if __name__ == '__main__':
    main()
```
